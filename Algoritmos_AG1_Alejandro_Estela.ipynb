```python
# Torres de Hanoi. Problema de divide y venceras. Divides el problemas en sub-problemas del mismo estílo y que sean más fáciles
def Torres_Hanoi(N, desde, hasta):
  # N numero de fichas, desde: nº poste desde, hasta: nº poste hasta
  if N == 1:
    print("Lleva la ficha desde: " + str(desde) + " hasta: " + str(hasta))
  else:
    Torres_Hanoi(N-1, desde, 6-desde-hasta)
    # Como calcular el pivote? Si me pasan el 1 y 3 el pivote es el 2. El valor restante
    print("Lleva la ficha desde: " + str(desde) + " hasta: " + str(hasta))
    Torres_Hanoi(N-1, 6-desde-hasta, hasta)


Torres_Hanoi(7,1,3)
```

```python
# Problema de cambio de monedas
def cambio_monedas(cantidad, sistema):
  quicksort(sistema, 0, len(sistema)-1) # Ordenar el sistema de monedas
  sistema.reverse()                     # Invertir las monedas de mayor a menor
  solucion = {}
  for v in sistema:
    monedas = cantidad // v             # Solo te quedas con el valor entero
    solucion[v] = monedas
    cantidad -= monedas * v
    if cantidad == 0: 
      return solucion
```

```python
# ALGORITMO QUICKSORT HOARE
def quicksort(lista, menor, mayor):
        if menor < mayor:
            pivote = partition(lista, menor, mayor)
            quicksort(lista, menor, pivote-1)
            quicksort(lista, pivote+1, mayor)

def partition(lista, menor, mayor):
        pivote = lista[menor]
        izq = menor + 1
        der = mayor
        while True:
          while izq <= der and lista[izq] <= pivote:
            izq += 1
          while izq <= der and lista[der] >= pivote:
            der -= 1
          if der < izq:
            break
          else:
            lista[izq], lista[der] = lista[der], lista[izq]
        lista[menor], lista[der] = lista[der], lista[menor]
        
        return der
```

```python
monedas = [5,1,10,20,100,50]
cambio = 125

cambio_monedas(cambio, monedas)
```
